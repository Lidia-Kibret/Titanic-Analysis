### Load Titanic Dataset
Load the Titanic dataset into a pandas DataFrame using `pd.read_csv`.

### Titanic Dataset — Data Cleaning
Handle missing values and remove unnecessary columns and duplicates.


```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
titanic = titanic.drop(columns=['Cabin'], errors='ignore')
titanic = titanic.drop_duplicates()

titanic.info()
```

### Survival Rate by Gender

This step calculates the **survival rate for each gender** in the Titanic dataset:

- The dataset is grouped by the `Sex` column.  
- For each gender group, the mean of the `Survived` column is calculated.  
- The resulting values represent the **proportion of passengers who survived** in each gender group, helping to identify which gender was more likely to survive.

```python
survival_gender = titanic.groupby('Sex')['Survived'].mean()
survival_gender
```

### Survival Rate by Passenger Class

This step calculates the **survival rate for each passenger class**:

- The dataset is grouped by the `Pclass` column (1st, 2nd, 3rd class).  
- For each class, the mean of the `Survived` column is computed.  
- The resulting values indicate the **proportion of passengers who survived in each class**, showing how class affected survival chances.

```python
survival_class = titanic.groupby('Pclass')['Survived'].mean()
survival_class
```

### Average Age per Passenger Class

This step calculates the **average age of passengers in each class**:

- The dataset is grouped by the `Pclass` column (1st, 2nd, 3rd class).  
- For each class, the mean of the `Age` column is computed.  
- The result shows the **typical age of passengers in each class**, helping to understand age distribution across classes.

```python
avg_age_class = titanic.groupby('Pclass')['Age'].mean()
avg_age_class
```

### Survival Rate by Age Group

This step categorizes passengers into **age groups** and calculates their survival rates:

1. **Age Grouping**:
   - `0-12` → Child  
   - `13-18` → Teen  
   - `19-35` → Young Adult  
   - `36-60` → Adult  
   - `61-100` → Senior  
   - `pd.cut` is used to assign each passenger to an age group.  

2. **Survival Calculation**:
   - The dataset is grouped by the `AgeGroup` column.  
   - For each group, the mean of the `Survived` column is computed.  
   - The result shows the **likelihood of survival for each age group**, indicating which ages were prioritized during evacuation.

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child','Teen','Young Adult','Adult','Senior']
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

survival_age_group = titanic.groupby('AgeGroup', observed=True)['Survived'].mean()
survival_age_group
```

### Female Passengers Who Survived

This step filters the dataset to identify **female passengers who survived**:

- The condition `(titanic['Sex']=='female')` selects all female passengers.  
- The condition `(titanic['Survived']==1)` selects those who survived.  
- Combining both conditions with `&` results in a **subset of female survivors**.  

This helps in analyzing **gender-based survival patterns** and confirms that females had higher survival rates compared to males.

```python
female_survived = titanic[(titanic['Sex']=='female') & (titanic['Survived']==1)]
female_survived
```

### Children Who Survived

This step filters the dataset to identify **child passengers who survived**:

- The condition `(titanic['Age'] <= 12)` selects all passengers aged 12 or below.  
- The condition `(titanic['Survived']==1)` selects those who survived.  
- Combining both conditions with `&` results in a **subset of children who survived**.  

This analysis helps determine if **children were prioritized for survival** during the Titanic disaster.

```python
children_survived = titanic[(titanic['Age'] <= 12) & (titanic['Survived']==1)]
children_survived
```

### First-Class Passengers Who Survived

This step filters the dataset to find **first-class passengers who survived**:

- The condition `(titanic['Pclass']==1)` selects passengers in **1st class**.  
- The condition `(titanic['Survived']==1)` selects passengers who **survived**.  
- Combining both conditions with `&` results in a **subset of first-class survivors**.  

This helps analyze whether **passenger class affected survival probability**.

```python
first_class_survived = titanic[(titanic['Pclass']==1) & (titanic['Survived']==1)]
first_class_survived
```

### Gender Most Likely to Survive

This step identifies **which gender had the highest survival rate**:

- `survival_gender` contains the **average survival rate** for each gender.  
- `idxmax()` returns the **index (gender)** corresponding to the **highest survival rate**.  

Result shows **females were more likely to survive** than males.

```python
most_likely_gender = survival_gender.idxmax()
most_likely_gender
```

### Passenger Class with Highest Survival Rate

This step determines **which passenger class had the highest chance of survival**:

- `survival_class` calculates the **average survival rate for each Pclass**.  
- `idxmax()` identifies the **class with the maximum survival rate**.  

Result indicates that **1st class passengers had the highest survival rate**.

```python
survival_class = titanic.groupby('Pclass')['Survived'].mean()
best_class = survival_class.idxmax()
best_class
```

### Survival Rate of Children

This step calculates the **survival rate specifically for children**:

- `survival_age_group` contains survival rates for all age groups.  
- Accessing `['Child']` gives the **proportion of children who survived**.

Result shows that **children had a higher survival rate compared to most other age groups**.

```python
child_survival_rate = survival_age_group['Child']
child_survival_rate
```

### Highest Survival Combination by Gender and Class

This step identifies which **gender and passenger class combination had the highest survival rate**:

- `groupby(['Sex','Pclass'])['Survived'].mean()` calculates the average survival rate for each combination of gender and class.  
- `.idxmax()` returns the combination with the **highest survival probability**.

The output shows the **most advantaged group on the Titanic in terms of survival**.

```python
highest_combination = titanic.groupby(['Sex','Pclass'])['Survived'].mean().idxmax()
highest_combination
```
